#Vanishing Gradient:

In machine learning the vanishing gradient problem is encountered when training ANN with gradient based learning method and back propogation in worst case this may completely stop the NN from further learning

##Condition:

1) when we have very smaller no 0.1x0.1x0.1x0.1==>0.0001

2) deep NN - lots of layers

3) if using sigmoid/tanh activation function (its sqeez the no)

#How to solve this problem

1) Reduce model complexity(not applicable) --> reducing the hidden layers

2) Using Relu activation function

3) proper weight initialization(gloroot/xavien --> tahh activation function
he --> Relu)

```python
import tensorflow as tf
import keras
from keras.layers import Dense
from keras.models import Sequential
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.datasets import make_moons
from sklearn.model_selection import train_test_split
```

```python
X,y=make_moons(n_samples=250,noise=0.05,random_state=42)
```

```python
X.shape, y.shape
```

```python
X
```

```python
y
```

```python
plt.scatter(X[:,0], X[:,1], c=y, s=100)
plt.show()
```

```python
model=Sequential()

model.add(Dense(10, activation='sigmoid', input_dim=2))
model.add(Dense(10, activation='sigmoid'))
model.add(Dense(10, activation='sigmoid'))
model.add(Dense(10, activation='sigmoid'))
model.add(Dense(10, activation='sigmoid'))
model.add(Dense(10, activation='sigmoid'))
model.add(Dense(10, activation='sigmoid'))
model.add(Dense(10, activation='sigmoid'))
model.add(Dense(10, activation='sigmoid'))
model.add(Dense(10, activation='sigmoid'))
model.add(Dense(10, activation='sigmoid'))
model.add(Dense(10, activation='sigmoid'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.get_weights()[0]
```

```python
old_weights = model.get_weights()[0]
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=42)
```

```python
model.fit(X_train, y_train, epochs=50)
```

```python
new_weights=model.get_weights()[0]
```

```python
old_weights
```

```python
new_weights
```

```python
percent_change=abs(100*(old_weights-new_weights)/old_weights)
percent_change
```

```python
from mlxtend.plotting import plot_decision_regions
plot_decision_regions(X_train,y_train.astype('int'),clf=model,legend=2)
```

#2) Using Relu activation function

```python
model2.compile(loss='binary_crossentropy', optimizer= 'adam',metrics=['accuracy'])
```

#3) proper weight initialization(gloroot/xavien --> tahh activation function he --> Relu)

```python
model3=Sequential()

model3.add(Dense(10,activation='relu',input_dim=2,kernel_initializer='he_normal'))
model3.add(Dense(10,activation='relu',kernel_initializer='he_normal'))
model3.add(Dense(10,activation='relu',kernel_initializer='he_normal'))
model3.add(Dense(10,activation='relu',kernel_initializer='he_normal'))
model3.add(Dense(1,activation='sigmoid'))
```

```python
model3.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
old_weight=model3.get_weights()[0]
```

```python
model3.fit(X_train,y_train,epochs=50)
```

```python
new_weight=model3.get_weights()[0]
```
